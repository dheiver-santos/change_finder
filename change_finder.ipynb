```python
from change_finder import ChangeFinder
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
f, (ax1, ax2) = plt.subplots(2, 1)
f.subplots_adjust(hspace=0.4)
points = np.concatenate([
    np.random.normal(1, 0.1, 200),
    np.random.normal(-1, 0.1, 200),
    np.random.normal(1, 0.1, 200),
])
ax1.plot(points)
ax1.set_title("data point")

cf = ChangeFinder()
scores = np.array([cf.update(p) for p in points])
diff = points - np.roll(points, 1)
ax2.plot(scores * diff)
ax2.set_title("anomaly score")
plt.savefig('foo.png')
```
